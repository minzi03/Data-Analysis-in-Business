```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
from pmdarima import auto_arima
```

```python
data = pd.read_csv("../../Dataset/AMV Historical Data.csv")

data['Price'] = data['Price'].str.replace(',', '').astype(float)
data['Date'] = pd.to_datetime(data['Date'])
data['Date']
```

```python
time_series = pd.DataFrame({'Date': data['Date'], 'Price': data['Price']})
time_series['Date'] = pd.to_datetime(time_series['Date'])
time_series['Price'] = time_series['Price'].astype(float)
time_series.set_index('Date', inplace=True)
time_series.sort_index(inplace=True)

time_series
```

```python
def evaluate_model(train_data, test_data, split_ratio):
    # Fit auto_arima model
    model = auto_arima(train_data, start_p=1, start_q=1,
                       max_p=5, max_q=5, m=12,
                       start_P=0, seasonal=False,
                       d=1, D=1, trace=True,
                       error_action='ignore',
                       suppress_warnings=True,
                       stepwise=True)

    order = model.get_params()['order']
    
    # Fit ARIMA model
    arima_model = ARIMA(train_data, order=order)
    arima_result = arima_model.fit()
    
    # Forecast on test set
    predictions = arima_result.predict(start=len(train_data), end=len(train_data)+len(test_data)-1, dynamic=False)
    
    # Forecast next 30 days
    forecast_30 = arima_result.forecast(steps=30)

    # Forecast 30-60 days
    forecast_60 = arima_result.get_forecast(steps=60).predicted_mean[30:60]

    # Forecast 60-90 days
    forecast_90 = arima_result.get_forecast(steps=90).predicted_mean[60:90]

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(train_data.index, train_data, label='Training Data')
    plt.plot(test_data.index, test_data, label='Actual Data')
    plt.plot(test_data.index, predictions, label='Predictions on Test')
    plt.plot(pd.date_range(test_data.index[-1], periods=30, freq='D'), forecast_30, label='Next 30 Days Forecast', color='red')
    plt.plot(pd.date_range(test_data.index[-1] + pd.Timedelta(days=30), periods=30, freq='D'), forecast_60, label='30-60 Days Forecast', color='purple')
    plt.plot(pd.date_range(test_data.index[-1] + pd.Timedelta(days=60), periods=30, freq='D'), forecast_90, label='60-90 Days Forecast', color='brown')

    plt.title(f'AMV Time Series Forecast ({split_ratio} Split)')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.show()

    # Convert to numpy arrays for metric calculations
    test_data = np.array(test_data['Price'])
    predictions = np.array(predictions)
    
    # Calculate RMSE
    test_rmse = np.sqrt(np.mean((test_data - predictions) ** 2))
    print('Testing RMSE:', test_rmse)

    # Calculate MAPE
    test_mape = np.mean(np.abs((test_data - predictions) / test_data)) * 100
    print('Testing MAPE:', test_mape)

    # Calculate MSLE
    test_msle = np.mean((np.log1p(test_data) - np.log1p(predictions)) ** 2)
    print('Testing MSLE:', test_msle)

# Example usage with different splits
# Assuming time_series is your DataFrame with 'Date' as index and 'Price' as column

# 8:2 split
train_data_8_2 = time_series.iloc[:int(0.8 * len(time_series))]
test_data_8_2 = time_series.iloc[int(0.8 * len(time_series)):]
print("8:2 split evaluation")
evaluate_model(train_data_8_2, test_data_8_2, "8:2")

# 7:3 split
train_data_7_3 = time_series.iloc[:int(0.7 * len(time_series))]
test_data_7_3 = time_series.iloc[int(0.7 * len(time_series)):]
print("7:3 split evaluation")
evaluate_model(train_data_7_3, test_data_7_3, "7:3")

# 9:1 split
train_data_9_1 = time_series.iloc[:int(0.9 * len(time_series))]
test_data_9_1 = time_series.iloc[int(0.9 * len(time_series)):]
print("9:1 split evaluation")
evaluate_model(train_data_9_1, test_data_9_1, "9:1")
```
